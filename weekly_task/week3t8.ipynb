```python
import pandas as pd
from google.colab import files
uploaded = files.upload()  # Upload your CSV file

# Get the filename of the uploaded file
filename = list(uploaded.keys())[0]
# Read the file using the correct filename and extension
df = pd.read_excel(filename) # Change read_csv to read_excel since user uploaded xlsx
df.head()

# Load the patient data
# Replace '/path/to/your/raw_patient_data.csv' with the actual path to your file
# df = pd.read_csv("raw_patient_data.csv") #This line was redundant

# Drop duplicate rows
df.drop_duplicates(inplace=True)

# Handle missing values (example: fill missing age with the median age)
df['Age'].fillna(df['Age'].median(), inplace=True)

# Convert date columns to datetime format
df['Admission Date'] = pd.to_datetime(df['Admission Date'], errors='coerce')
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'], errors='coerce')

# Normalize text data (example: standardize gender values)
df['Gender'] = df['Gender'].str.strip().str.capitalize()

# Save the cleaned data to a CSV file
df.to_csv("cleaned_patient_data.csv", index=False)

print("Cleaned data has been saved as 'cleaned_patient_data.csv'")
```
